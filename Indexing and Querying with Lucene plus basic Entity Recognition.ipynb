```python
import subprocess
import time
import spacy
import en_core_web_sm

nlp = spacy.load('en_core_web_sm', vectors=False)

PATH = "/Users/federico/Google Drive/Courses/Master BIG DATA/Social Media-Basili/Final Project/BigData_2018-2019_TM_SMA_project/search_engine/"
# Oracle index directory
index_dir=PATH+"tmp_index"
#Predicted index directory
index_dirpred=PATH+"index_last"
#query = "movie ghost rider"

query = input('Insert the query:  ')

max_results = 20

target_polarity= str(input("inserisci polarity desiderata tra 'positive', 'negative' e  'neutral':  "))

out, err = subprocess.Popen(["java", "-cp", "lib/lucene-core-3.6.0.jar:.", "it.uniroma2.sag.assignment.ir.lucene.twitter.search.SimpleTweetSearch", 
                             index_dirpred, query, str(max_results)], stdout=subprocess.PIPE).communicate() 

results = out.splitlines()
res_id = 1 

for result in results:
    split = result.decode("utf-8").split("\t");
    score= split[0]
    polarity = split[1]
    tweet = split[2]
    hashtags = split[3].split(" ")

    if polarity== target_polarity:
        print("Document "+str(res_id))
        print(score)
        print(polarity)
        print(tweet)
        print(hashtags)
        print()
        res_id = res_id + 1
        doc = nlp(u'{}'.format(tweet))
        # check name entities
        if doc.ents:
            for ent in doc.ents:
                print('Entity found \t:',ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
                print()
        else:
            print('No named entities found.')          
                     
        
print('\n\n The % of {} tweets retrived over the first {} documents is {}'.format(target_polarity, max_results, str(res_id*100/max_results)+'%'))
print('The absolute number of {} tweets retrived over the first {} documents is {}'.format(target_polarity, max_results, res_id))

```
